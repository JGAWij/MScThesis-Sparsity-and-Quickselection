```python
import os
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from sklearn import preprocessing
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
X_train = X_train.astype('float32')
X_test  = X_test.astype('float32')

X_train = X_train.reshape((X_train.shape[0],X_train.shape[1]*X_train.shape[2]))
X_test  = X_test.reshape((X_test.shape[0],X_test.shape[1]*X_test.shape[2]))

scaler = preprocessing.StandardScaler().fit(X_train)

# Get the number of samples
num_samples = X_train.shape[0]

# Define the subsample ratio
subsample_ratio = 0.5

# Compute the number of samples in the subsample
num_subsamples = int(np.ceil(num_samples * subsample_ratio))

# Randomly select the indices of the subsample
subsample_indices = np.random.choice(num_samples, size=num_subsamples, replace=False)

# Create the subsample
X_train_subsample = X_train[subsample_indices]
y_train_subsample = y_train[subsample_indices]

# Save the subsample with the same name as the original data
np.save(os.path.join(os.getcwd(), 'X_train.npy'), X_train_subsample)
np.save(os.path.join(os.getcwd(), 'y_train.npy'), y_train_subsample)

# Load the subsample
X_train = np.load(os.path.join(os.getcwd(), 'X_train.npy'))
y_train = np.load(os.path.join(os.getcwd(), 'y_train.npy'))

print(X_train.shape[0])
print(y_train.shape[0])
```

```python
import numpy as np
import random
from sklearn.preprocessing import label_binarize

def adjust_mnist2(x, y, target_gini, margin):
    # Calculate Gini coefficient and plot Lorenz curve
    class_prop_sorted = np.sort(np.bincount(y) / len(y))[::-1]
    lorenz_x, cumulative_prop = np.linspace(0, 1, len(class_prop_sorted)), np.cumsum(class_prop_sorted)
    lorenz_y, area_lorenz = cumulative_prop / cumulative_prop[-1], np.trapz(cumulative_prop, lorenz_x)
    area_eq = 0.5
    gini_coeff = (area_eq - area_lorenz) / area_eq
    plt.plot(lorenz_x, lorenz_y, label='Lorenz curve'); plt.plot([0, 1], [0, 1], label='Perfect equality')
    plt.title('Lorenz curve for original MNIST dataset'); plt.xlabel('Proportion of classes')
    plt.ylabel('Proportion of examples'); plt.legend(); plt.show()
    print(f'Gini coefficient: {gini_coeff}')
    
    # Define selected classes and calculate their proportion of data
    y = label_binarize(y, classes=np.arange(10))
    selected_classes = random.sample(range(10), 4)
    class_proportions = y[:, selected_classes].sum(axis=0) / y.sum()
    
    # Calculate target number of samples to remove from each selected class
    target_removals = np.maximum(((class_proportions - target_gini * 0.1) * y.sum(axis=0)[selected_classes]).astype(int), 0)
    print("Target Removals", target_removals)
    
    # Loop until target gini coefficient is achieved
    while True:
        y_selected = y[np.isin(np.argmax(y, axis=1), selected_classes)]
        if len(y_selected) == 0 or not all(np.isin(selected_classes, np.argmax(y, axis=1))):
            print("no samples left in selected classes or not all selected classes are present in data")
            return None
        y_removed = y.copy()
        for c in selected_classes:
            c_indices = np.where(y_removed[:, c] == 1)[0]
            if len(c_indices) == 0:
                print("selected class not present in data")
                return None
            n_remove = target_removals[c]
            if n_remove > 0:
                y_removed[random.sample(c_indices, n_remove), :] = 0
        x_adjusted, y_adjusted = x[y_removed.any(axis=1)], y_removed[y_removed.any(axis=1)]
        class_prop_sorted = np.sort(np.sum(label_binarize(y_adjusted, classes=np.arange(10)), axis=0) / len(y_adjusted))[::-1]
        lorenz_x, cumulative_prop = np.linspace(0, 1, len(class_prop_sorted)), np.cumsum(class_prop_sorted)
        lorenz_y, area_lorenz = cumulative_prop / cumulative_prop[-1], np.trapz(lorenz_y, lorenz_x)
        current_gini = (area_eq - area_lorenz) / area_eq
        print("current_gini", current_gini)
        if abs(current_gini - target_gini) <= margin:
            return x_adjusted, y_adjusted
```

```python
import os
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from sklearn import preprocessing
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
X_train = X_train.astype('float32')
X_test  = X_test.astype('float32')

X_train = X_train.reshape((X_train.shape[0],X_train.shape[1]*X_train.shape[2]))
X_test  = X_test.reshape((X_test.shape[0],X_test.shape[1]*X_test.shape[2]))

scaler = preprocessing.StandardScaler().fit(X_train)

print(X_train.shape[0])
# Get the number of samples
num_samples = X_train.shape[0]

# Define the subsample ratio
subsample_ratio = 0.5

# Compute the number of samples in the subsample
num_subsamples = int(np.ceil(num_samples * subsample_ratio))

# Randomly select the indices of the subsample
subsample_indices = np.random.choice(num_samples, size=num_subsamples, replace=False)

# Create the subsample
X_train_subsample = X_train[subsample_indices]
y_train_subsample = y_train[subsample_indices]

# Save the subsample with the same name as the original data
np.save(os.path.join(os.getcwd(), 'X_train.npy'), X_train_subsample)
np.save(os.path.join(os.getcwd(), 'y_train.npy'), y_train_subsample)

# Load the subsample
X_train = np.load(os.path.join(os.getcwd(), 'X_train.npy'))
y_train = np.load(os.path.join(os.getcwd(), 'y_train.npy'))

print(X_train.shape[0])
print(y_train.shape[0])
```

```python
import numpy as np
import random
from sklearn.preprocessing import label_binarize

def adjust_mnist2(x, y, target_gini, margin):
    print("Y before:", y)
    class_freq = np.bincount(y)
    class_prop = class_freq / len(y)
    class_prop_sorted = np.sort(class_prop)[::-1]
    cumulative_prop = np.cumsum(class_prop_sorted)
    lorenz_x = np.linspace(0, 1, len(class_prop_sorted))
    lorenz_y = np.cumsum(class_prop_sorted) / cumulative_prop[-1]
    area_lorenz = np.trapz(lorenz_y, lorenz_x)
    area_eq = 0.5
    gini_coeff = (area_eq - area_lorenz) / area_eq

    # Plot the Lorenz curve
    plt.plot(lorenz_x, lorenz_y, label='Lorenz curve')
    plt.plot([0, 1], [0, 1], label='Perfect equality')
    plt.title('Lorenz curve for original MNIST dataset')
    plt.xlabel('Proportion of classes')
    plt.ylabel('Proportion of examples')
    plt.legend()
    plt.show()
    print(f'Gini coefficient: {gini_coeff}')
    
    # Define selected classes and calculate proportion of data in each class
    y = label_binarize(y, classes=np.arange(10))
    print("Print y:", y)
    selected_classes = random.sample(list(range(10)), 4)
    print("selected_classes", selected_classes)
    class_proportions = y[:, selected_classes].sum(axis=0) / y.sum()
    print("Clas proportions:", class_proportions)
    
    # Initialize array to keep track of removed data
    target_removals = np.zeros(y.shape[1], dtype=int)

    # Calculate target number of samples to remove from each selected class
    scale_factor = 0.1
    target_removals[selected_classes] = ((class_proportions - target_gini * scale_factor) * y.sum(axis=0)[selected_classes]).astype(int)
    target_removals = np.maximum(target_removals, 0)  
    # Ensure non-negative values
    print("Target Removals", target_removals)
    
    # Initialize array to keep track of removed data
    removed_data = np.zeros(y.shape, dtype=bool)
    
    # Loop until target gini coefficient is achieved
    while True:
            y_selected = y[np.isin(np.argmax(y, axis=1), selected_classes)]
            print ("y_selected", y_selected)
            if len(y_selected) == 0:
                print("no samples left in selected classes")
                return None # no samples left in selected classes
            if not all(np.isin(selected_classes, np.argmax(y, axis=1))):
                print("# not all selected classes are present in data")
                return None
            
            #REMOVAL PART
            y_removed = y.copy()
            for c in selected_classes:
                c_indices = np.where(y_removed[:, c] == 1)[0] # select indices for class c
                if len(c_indices) == 0:
                    print("selected class not present in data")
                    return None # selected class not present in data
                n_remove = target_removals[c]
                if n_remove > 0:
                    indices_to_remove = random.sample(list(c_indices), n_remove) # randomly select samples to remove
                    y_removed[indices_to_remove, :] = 0 # mark samples to be removed
        
            x_adjusted = x[y_removed.any(axis=1)]
            y_adjusted = y_removed[y_removed.any(axis=1)]
            y_binarized = label_binarize(y_adjusted, classes=np.arange(10))
            class_freq = np.sum(y_binarized, axis=0)
            class_prop = class_freq / len(y_binarized)
            class_prop_sorted = np.sort(class_prop)[::-1]
            cumulative_prop = np.cumsum(class_prop_sorted)
            lorenz_x = np.linspace(0, 1, len(class_prop_sorted))
            lorenz_y = np.cumsum(class_prop_sorted) / cumulative_prop[-1]
            area_lorenz = np.trapz(lorenz_y, lorenz_x)
            area_eq = 0.5
            current_gini = (area_eq - area_lorenz) / area_eq
            print("current_gini", current_gini)
            if abs(current_gini - target_gini) <= margin:
                return x_adjusted, y_adjusted
```

```python
# Assuming X_train and y_train are the training data and labels of the MNIST dataset
target_gini = 0.6
margin= 0.05
adjusted_X_train, adjusted_y_train = adjust_mnist2(X_train, y_train, target_gini, margin)
```
